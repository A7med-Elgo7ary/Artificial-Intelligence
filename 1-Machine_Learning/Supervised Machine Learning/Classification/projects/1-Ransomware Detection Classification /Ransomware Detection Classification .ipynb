```python
!pip install pefile
import hashlib
import os
import pefile


def extract_features(file_path):
    """
    Extracts all 56  static features from a PE file.
    :param file_path: Path to the PE file.
    :return: A dictionary containing all features.
    """
    features = {}

    # Feature 0: MD5 hash
    with open(file_path, "rb") as f:
        file_data = f.read()
        features['md5'] = hashlib.md5(file_data).hexdigest()

    # Feature 1: SHA-1 hash
    features['sha1'] = hashlib.sha1(file_data).hexdigest()

    # Feature 2: File extension
    features['file_extension'] = os.path.splitext(file_path)[1]

    # Load the PE file
    pe = pefile.PE(file_path)

    # Feature 3: EntryPoint
    features['EntryPoint'] = pe.OPTIONAL_HEADER.AddressOfEntryPoint

    # Feature 4: PEType
    features['PEType'] = pe.FILE_HEADER.Machine

    # Feature 5: MachineType
    features['MachineType'] = pe.FILE_HEADER.Machine

    # Feature 6: Magic number
    features['magic_number'] = pe.DOS_HEADER.e_magic

    # Feature 7: Bytes on last page
    features['bytes_on_last_page'] = pe.DOS_HEADER.e_cblp

    # Feature 8: Pages in file
    features['pages_in_file'] = pe.DOS_HEADER.e_cp

    # Feature 9: Relocations
    features['relocations'] = pe.FILE_HEADER.NumberOfRelocations

    # Feature 10: Size of header
    features['size_of_header'] = pe.OPTIONAL_HEADER.SizeOfHeaders

    # Feature 11: Min extra paragraphs
    features['min_extra_paragraphs'] = pe.DOS_HEADER.e_minalloc

    # Feature 12: Max extra paragraphs
    features['max_extra_paragraphs'] = pe.DOS_HEADER.e_maxalloc

    # Feature 13: Initial SS value
    features['init_ss_value'] = pe.DOS_HEADER.e_ss

    # Feature 14: Initial SP value
    features['init_sp_value'] = pe.DOS_HEADER.e_sp

    # Feature 15: Initial IP value
    features['init_ip_value'] = pe.DOS_HEADER.e_ip

    # Feature 16: Initial CS value
    features['init_cs_value'] = pe.DOS_HEADER.e_cs

    # Feature 17: Overlay number
    features['over_lay_number'] = pe.get_overlay()

    # Feature 18: OEM identifier
    features['oem_identifier'] = pe.DOS_HEADER.e_oemid

    # Feature 19: Address of NE header
    features['address_of_ne_header'] = pe.DOS_HEADER.e_lfanew

    # Feature 20: Magic
    features['Magic'] = pe.OPTIONAL_HEADER.Magic

    # Feature 21: SizeOfCode
    features['SizeOfCode'] = pe.OPTIONAL_HEADER.SizeOfCode

    # Feature 22: SizeOfInitializedData
    features['SizeOfInitializedData'] = pe.OPTIONAL_HEADER.SizeOfInitializedData

    # Feature 23: SizeOfUninitializedData
    features['SizeOfUninitializedData'] = pe.OPTIONAL_HEADER.SizeOfUninitializedData

    # Feature 24: AddressOfEntryPoint
    features['AddressOfEntryPoint'] = pe.OPTIONAL_HEADER.AddressOfEntryPoint

    # Feature 25: BaseOfCode
    features['BaseOfCode'] = pe.OPTIONAL_HEADER.BaseOfCode

    # Feature 26: BaseOfData
    features['BaseOfData'] = pe.OPTIONAL_HEADER.BaseOfData

    # Feature 27: ImageBase
    features['ImageBase'] = pe.OPTIONAL_HEADER.ImageBase

    # Feature 28: SectionAlignment
    features['SectionAlignment'] = pe.OPTIONAL_HEADER.SectionAlignment

    # Feature 29: FileAlignment
    features['FileAlignment'] = pe.OPTIONAL_HEADER.FileAlignment

    # Feature 30: OperatingSystemVersion
    features['OperatingSystemVersion'] = pe.OPTIONAL_HEADER.MajorOperatingSystemVersion

    # Feature 31: ImageVersion
    features['ImageVersion'] = pe.OPTIONAL_HEADER.MajorImageVersion

    # Feature 32: SizeOfImage
    features['SizeOfImage'] = pe.OPTIONAL_HEADER.SizeOfImage

    # Feature 33: SizeOfHeaders
    features['SizeOfHeaders'] = pe.OPTIONAL_HEADER.SizeOfHeaders

    # Feature 34: Checksum
    features['Checksum'] = pe.OPTIONAL_HEADER.CheckSum

    # Feature 35: Subsystem
    features['Subsystem'] = pe.OPTIONAL_HEADER.Subsystem

    # Feature 36: DllCharacteristics
    features['DllCharacteristics'] = pe.OPTIONAL_HEADER.DllCharacteristics

    # Feature 37: SizeofStackReserve
    features['SizeofStackReserve'] = pe.OPTIONAL_HEADER.SizeOfStackReserve

    # Feature 38: SizeofStackCommit
    features['SizeofStackCommit'] = pe.OPTIONAL_HEADER.SizeOfStackCommit

    # Feature 39: SizeofHeapCommit
    features['SizeofHeapCommit'] = pe.OPTIONAL_HEADER.SizeOfHeapCommit

    # Feature 40: SizeofHeapReserve
    features['SizeofHeapReserve'] = pe.OPTIONAL_HEADER.SizeOfHeapReserve

    # Feature 41: LoaderFlags
    features['LoaderFlags'] = pe.OPTIONAL_HEADER.LoaderFlags

    # Extract section-specific features
    for section in pe.sections:
        section_name = section.Name.decode().strip('\x00')
        if section_name == ".text":
            # Feature 42: .text VirtualSize
            features['text_VirtualSize'] = section.Misc_VirtualSize
            # Feature 43: .text VirtualAddress
            features['text_VirtualAddress'] = section.VirtualAddress
            # Feature 44: .text SizeOfRawData
            features['text_SizeOfRawData'] = section.SizeOfRawData
            # Feature 45: .text PointerToRawData
            features['text_PointerToRawData'] = section.PointerToRawData
            # Feature 46: .text PointerToRelocations
            features['text_PointerToRelocations'] = section.PointerToRelocations
            # Feature 47: .text PointerToLineNumbers
            features['text_PointerToLineNumbers'] = section.PointerToLinenumbers
            # Feature 48: .text Characteristics
            features['text_Characteristics'] = section.Characteristics
        elif section_name == ".rdata":
            # Feature 49: .rdata VirtualSize
            features['rdata_VirtualSize'] = section.Misc_VirtualSize
            # Feature 50: .rdata VirtualAddress
            features['rdata_VirtualAddress'] = section.VirtualAddress
            # Feature 51: .rdata SizeOfRawData
            features['rdata_SizeOfRawData'] = section.SizeOfRawData
            # Feature 52: .rdata PointerToRawData
            features['rdata_PointerToRawData'] = section.PointerToRawData
            # Feature 53: .rdata PointerToRelocations
            features['rdata_PointerToRelocations'] = section.PointerToRelocations
            # Feature 54: .rdata PointerToLineNumbers
            features['rdata_PointerToLineNumbers'] = section.PointerToLinenumbers
            # Feature 55: .rdata Characteristics
            features['rdata_Characteristics'] = section.Characteristics

    # Close the PE file
    pe.close()

    return features


# Example usage
if __name__ == "__main__":
    file_path = "file.exe"  # Replace with your file path
    features = extract_features(file_path)
    for key, value in features.items():
        print(f"{key}: {value}")
```
