```python
# Filling the missing float values with median
df['MinTemp']=df['MinTemp'].fillna(df['MinTemp'].median())
df['MaxTemp']=df['MaxTemp'].fillna(df['MaxTemp'].median())
df['Rainfall']=df['Rainfall'].fillna(df['Rainfall'].median())
df['Evaporation']=df['Evaporation'].fillna(df['Evaporation'].median())
df['Sunshine']=df['Sunshine'].fillna(df['Sunshine'].median())
df['WindGustSpeed']=df['WindGustSpeed'].fillna(df['WindGustSpeed'].median())
df['WindSpeed9am']=df['WindSpeed9am'].fillna(df['WindSpeed9am'].median())
df['WindSpeed3pm']=df['WindSpeed3pm'].fillna(df['WindSpeed3pm'].median())
df['Humidity9am']=df['Humidity9am'].fillna(df['Humidity9am'].median())
df['Humidity3pm']=df['Humidity3pm'].fillna(df['Humidity3pm'].median())
df['Pressure9am']=df['Pressure9am'].fillna(df['Pressure9am'].median())
df['Pressure3pm']=df['Pressure3pm'].fillna(df['Pressure3pm'].median())
df['Cloud9am']=df['Cloud9am'].fillna(df['Cloud9am'].median())
df['Cloud3pm']=df['Cloud3pm'].fillna(df['Cloud3pm'].median())
df['Temp9am']=df['Temp9am'].fillna(df['Temp9am'].median())
df['Temp3pm']=df['Temp3pm'].fillna(df['Temp3pm'].median())
```

```python
# Filling the missing object values with mode
df['WindGustDir']=df['WindGustDir'].fillna(df['WindGustDir'].mode()[0])
df['WindDir9am']=df['WindDir9am'].fillna(df['WindDir9am'].mode()[0])
df['WindDir3pm']=df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0])
df['RainToday']=df['RainToday'].fillna(df['RainToday'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Date']=lb.fit_transform(df['Date'])
df['Location']=lb.fit_transform(df['Location'])
df['WindGustDir']=lb.fit_transform(df['WindGustDir'])
df['WindDir9am']=lb.fit_transform(df['WindDir9am'])
df['WindDir3pm']=lb.fit_transform(df['WindDir3pm'])
df['RainToday']=lb.fit_transform(df['RainToday'])
```

## **Splitting into Training and Testing Data**

```python
# Separating the input and output features
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

## **Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train_normalized=scaler.transform(x_train)
x_test_normalized=scaler.transform(x_test)
```

## **BALANCING**

```python


from imblearn.over_sampling import RandomOverSampler
from sklearn.datasets import make_classification

over_sampler = RandomOverSampler(random_state=42)
x_train_balanced, y_train_balanced = over_sampler.fit_resample(x_train_normalized, y_train)
```

```python
sns.countplot(x=y_train_balanced)
```

```python
# Balancing by Resampling Technique
from imblearn.under_sampling import RandomUnderSampler
from sklearn.datasets import make_classification
under_sampler = RandomUnderSampler(random_state=42)
x_train_balanced,y_train_balanced = under_sampler.fit_resample(x_train_normalized,y_train)
```

```python
sns.countplot(x=y_train_balanced)
```

## **FEATURE SELECTION**

```python
# Chi-squared test
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
selector=SelectKBest(chi2, k=18)
x_train_selected=selector.fit_transform(x_train_balanced,y_train_balanced)
x_test_selected=selector.transform(x_test_normalized)
# To get the indices of the selected features
selected_feature_indices=selector.get_support(indices=True)
# To get the names of the selected features
selected_feature_names=x.columns[selected_feature_indices]
# To print the selected feature names
print("Selected Features:")
print(selected_feature_names)
```

## **Model creation and Performance evaluation**

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier


Logistic = LogisticRegression(solver='lbfgs', max_iter=200)
naive=GaussianNB()
decision=DecisionTreeClassifier(criterion='entropy')
random=RandomForestClassifier(n_estimators=100)
knn=KNeighborsClassifier(n_neighbors=7)
vector=SVC()
Gradient_Boosting= GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
```

## **Find Optimal K for KNN**

```python
from sklearn.model_selection import cross_val_score
k_values = range(1, 21)
scores = []

# Loop through the K values to evaluate performance
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    cv_scores = cross_val_score(knn, x, y, cv=5, scoring='accuracy')  # 5-fold cross-validation
    scores.append(cv_scores.mean())

# Find the K value with the highest cross-validation score
optimal_k = k_values[np.argmax(scores)]
print(f"The optimal K for KNN is {optimal_k}")
```

```python
k_values = range(1, 21)
error_rates = []

# Loop through the K values to evaluate performance
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train_selected, y_train_balanced)
    y_pred = knn.predict(x_test_selected)
    error_rate = 1 - accuracy_score(y_test, y_pred)
    error_rates.append(error_rate)
# Plot the error rates against K values
plt.plot(k_values, error_rates, 'bx-')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Error Rate')
plt.title('Elbow Method for Optimal K')
plt.show()
```

## **training and testing all models**

```python

lst_models=[Logistic,naive,decision,random,knn,vector,Gradient_Boosting]
accuracy=[]
for i in lst_models:
    print("model: " + str(i))
    i.fit(x_train_selected,y_train_balanced)
    y_pred=i.predict(x_test_selected)
    score=accuracy_score(y_test,y_pred)
    accuracy.append(score)
    cm = confusion_matrix(y_test, y_pred)
    AccuracyScore = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    print("confusion_matrix:\n" + str(cm))
    print(f"Accuracy Score: {AccuracyScore}")
    print(f"Classification Report:\n{report}")
    print("-------------------------")
```

```python
# Function to predict new values
def predict_with_model(model, new_x_test):
    predictions = model.predict(new_x_test)
    return predictions
```
